```python
def myfun(n):
    if n > 100:
        return 'very large'
    elif n > 50:
        return 'large'
    elif n > 20:
        return 'medium'
    elif n > 0:
        return 'positive small'
    elif n == 0:
        return 'zero'
    elif n > -10:
        return 'negative small'
    elif n > -50:
        return 'negative medium'
    elif n > -100:
        return 'negative large'
    else:
        return 'very negative'

print(myfun(10000))

print(myfun(87))

print(myfun(43))

print(myfun(2))

print(myfun(0))

print(myfun(-11))

print(myfun(-51))

print(myfun(-101))

print(myfun(150))
```
